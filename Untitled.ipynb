```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes

# 1. Création d'une clé unique pour tous les tests (32 octets = 256 bits)
key = get_random_bytes(32)

# 2. Le message à chiffrer (doit être en bytes)
message = b"Ceci est un message secret pour tester les modes AES."

print("Message original :", message)
print("Clé utilisée     :", key.hex())
```

```python
print("\n--- TEST DU MODE CBC ---")

# 1. Générer un IV aléatoire (16 octets pour AES)
iv_cbc = get_random_bytes(16)

# 2. Chiffrement
cipher_cbc_enc = AES.new(key, AES.MODE_CBC, iv_cbc)

# Le padding est obligatoire car la taille du message n'est pas multiple de 16
padde = pad(message, AES.block_size)
ciphertext_cbc = cipher_cbc_enc.encrypt(padde)

print("Chiffré (CBC) :", ciphertext_cbc.hex())

# 3. Déchiffrement
#On doit fournir le MÊME IV qu'au chiffrement
cipher_cbc_dec = AES.new(key, AES.MODE_CBC, iv_cbc)
decrypted_padde = cipher_cbc_dec.decrypt(ciphertext_cbc)
decrypted_cbc = unpad(decrypted_padde, AES.block_size)

print("Déchiffré     :", decrypted_cbc)
```

```python
print("\n--- TEST DU MODE ECB ---")

# 1. Chiffrement 
cipher_ecb_enc = AES.new(key, AES.MODE_ECB)
# Le padding reste obligatoire
msg_padde = pad(message, AES.block_size)
ciphertext_ecb = cipher_ecb_enc.encrypt(msg_padde)

print("Chiffré (ECB) :", ciphertext_ecb.hex())

# 2. Déchiffrement
cipher_ecb_dec = AES.new(key, AES.MODE_ECB)
decrypted_padde = cipher_ecb_dec.decrypt(ciphertext_ecb)
decrypted_ecb = unpad(decrypted_padde, AES.block_size)

print("Déchiffré     :", decrypted_ecb)
```

```python
print("\n--- TEST DU MODE OFB ---")

# 1. Générer un  IV 
iv_ofb = get_random_bytes(16)

# 2. Chiffrement
cipher_ofb_enc = AES.new(key, AES.MODE_OFB, iv_ofb)
# Bien que OFB fonctionne comme un flot, on utilise souvent le padding 
# par habitude avec AES, ou on s'assure que le message est géré correctement.
msg_padde = pad(message, AES.block_size) 
ciphertext_ofb = cipher_ofb_enc.encrypt(msg_padde)

print("Chiffré (OFB) :", ciphertext_ofb.hex())

# 3. Déchiffrement
cipher_ofb_dec = AES.new(key, AES.MODE_OFB, iv_ofb)
decrypted_padde = cipher_ofb_dec.decrypt(ciphertext_ofb)
decrypted_ofb = unpad(decrypted_padde, AES.block_size)

print("Déchiffré     :", decrypted_ofb)
```

```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad
from Crypto.Random import get_random_bytes
import os

def chiffrer_image(nom_fichier_entree, mode_aes, nom_mode_str):
    # 1. Lecture du fichier image binaire
    with open(nom_fichier_entree, 'rb') as f:
        data = f.read()

    # 2. Séparation En-tête (Header) / Données (Pixels)
    # Pour un BMP standard, l'en-tête fait 54 octets.
    # On garde l'en-tête intact pour que l'OS reconnaisse l'image.
    header = data[:54]
    body = data[54:]

    # 3. Préparation de la clé et IV
    key = b'0123456789ABCDEF0123456789ABCDEF' # Clé statique pour l'exemple (32 bytes)
    iv = b'0123456789ABCDEF'                  # IV statique (16 bytes)

    # 4. Configuration du mode
    if mode_aes == AES.MODE_ECB:
        cipher = AES.new(key, mode_aes) # Pas d'IV pour ECB
    else:
        cipher = AES.new(key, mode_aes, iv)

    # 5. Chiffrement du corps de l'image
    # On doit padder le corps car sa taille n'est peut-être pas multiple de 16
    body_padded = pad(body, AES.block_size)
    body_chiffre = cipher.encrypt(body_padded)

    # 6. Reconstruction : Header original + Corps chiffré
    # (On coupe le surplus de padding pour garder une taille proche, 
    # sinon certains visualiseurs d'images râlent, mais pour ce TP c'est facultatif)
    data_finale = header + body_chiffre[:len(body)] 

    # 7. Écriture du fichier de sortie
    nom_sortie = f"output_{nom_mode_str}.bmp"
    with open(nom_sortie, 'wb') as f:
        f.write(data_finale)
    
    print(f"Image générée : {nom_sortie}")

# --- EXÉCUTION ---

# Assurez-vous d'avoir un fichier 'input.bmp' dans le dossier !
try:
    print("--- Génération des images chiffrées ---")
    
    # Test ECB (Vous devriez voir les formes de l'image originale)
    chiffrer_image('oiseaux.bmp', AES.MODE_ECB, "ECB")
    
    # Test CBC (Doit ressembler à de la neige / bruit TV)
    chiffrer_image('oiseaux.bmp', AES.MODE_CBC, "CBC")
    
    # Test OFB (Doit ressembler à de la neige / bruit TV)
    chiffrer_image('oiseaux.bmp', AES.MODE_OFB, "OFB")
    
    print("\nTerminé ! Ouvrez les images output_*.bmp pour comparer.")

except FileNotFoundError:
    print("ERREUR : Le fichier 'oiseaux.bmp' est introuvable.")
    print("Veuillez placer une image BMP dans le dossier du script.")
```

```python
import os
import hashlib
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes

# --- FONCTIONS UTILITAIRES ---

def generer_cle_depuis_mdp(mot_de_passe):
    """Transforme un mot de passe texte en clé de 32 octets (AES-256) via SHA-256."""
    digest = hashlib.sha256(mot_de_passe.encode('utf-8')).digest()
    return digest

def chiffrer_fichier(nom_fichier, mot_de_passe):
    """Chiffre un fichier et ajoute l'extension .enc"""
    try:
        # 1. Préparation de la clé et de l'IV
        key = generer_cle_depuis_mdp(mot_de_passe)
        iv = get_random_bytes(16) # IV aléatoire indispensable pour CBC
        
        # 2. Lecture du fichier original
        with open(nom_fichier, 'rb') as f:
            data_originale = f.read()
            
        # 3. Chiffrement (AES-CBC + Padding)
        cipher = AES.new(key, AES.MODE_CBC, iv)
        data_chiffree = cipher.encrypt(pad(data_originale, AES.block_size))
        
        # 4. Écriture : On sauvegarde l'IV *suivi* des données chiffrées
        nom_sortie = nom_fichier + ".enc"
        with open(nom_sortie, 'wb') as f:
            f.write(iv)             # Les 16 premiers octets sont l'IV
            f.write(data_chiffree)  # Le reste est le message
            
        print(f" Succès ! Fichier chiffré créé : {nom_sortie}")
        return nom_sortie

    except FileNotFoundError:
        print(f" Erreur : Le fichier '{nom_fichier}' est introuvable.")
    except Exception as e:
        print(f" Erreur inattendue : {e}")

def dechiffrer_fichier(nom_fichier_chiffre, mot_de_passe):
    """Déchiffre un fichier .enc et retire l'extension"""
    try:
        # 1. Génération de la clé
        key = generer_cle_depuis_mdp(mot_de_passe)
        
        # 2. Lecture du fichier chiffré
        with open(nom_fichier_chiffre, 'rb') as f:
            iv = f.read(16)        # On récupère l'IV (les 16 premiers octets)
            data_chiffree = f.read() # Le reste
            
        # 3. Déchiffrement
        cipher = AES.new(key, AES.MODE_CBC, iv)
        try:
            data_dechiffree = unpad(cipher.decrypt(data_chiffree), AES.block_size)
        except ValueError:
            print(" Erreur : Mot de passe incorrect ou fichier corrompu (Padding invalide).")
            return

        # 4. Écriture du fichier déchiffré
        # On enlève ".enc" et on ajoute "decrypted_" pour ne pas écraser l'original
        nom_base = os.path.splitext(nom_fichier_chiffre)[0]
        nom_sortie = "decrypted_" + nom_base
        
        with open(nom_sortie, 'wb') as f:
            f.write(data_dechiffree)
            
        print(f" Succès ! Fichier déchiffré : {nom_sortie}")
        return nom_sortie

    except FileNotFoundError:
        print(f" Erreur : Le fichier '{nom_fichier_chiffre}' est introuvable.")

# --- INTERFACE UTILISATEUR (MENU) ---

def main():
    print("=== APPLICATION DE CHIFFREMENT AES-256 ===")
    print("1. Chiffrer un fichier")
    print("2. Déchiffrer un fichier")
    choix = input("Votre choix (1 ou 2) : ")

    if choix == '1':
        fichier = input("Nom du fichier à chiffrer (ex: image.bmp) : ")
        mdp = input("Choisissez un mot de passe : ")
        chiffrer_fichier(fichier, mdp)
        
    elif choix == '2':
        fichier = input("Nom du fichier chiffré (ex: image.bmp.enc) : ")
        mdp = input("Entrez le mot de passe original : ")
        dechiffrer_fichier(fichier, mdp)
        
    else:
        print("Choix invalide.")

# Pour lancer l'application directement dans le notebook/console :
if __name__ == "__main__":
    # On crée un petit fichier test pour la démo si besoin
    if not os.path.exists("test_secret.txt"):
        with open("test_secret.txt", "w") as f:
            f.write("Ceci est une donnée confidentielle pour le TP Crypto.")
    
    main()
```

```python
import os
import hashlib
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes

# --- FONCTIONS UTILITAIRES ---

def generer_cle_depuis_mdp(mot_de_passe):
    """Transforme un mot de passe texte en clé de 32 octets (AES-256) via SHA-256."""
    digest = hashlib.sha256(mot_de_passe.encode('utf-8')).digest()
    return digest

def chiffrer_fichier(nom_fichier, mot_de_passe):
    """Chiffre un fichier et ajoute l'extension .enc"""
    try:
        # 1. Préparation de la clé et de l'IV
        key = generer_cle_depuis_mdp(mot_de_passe)
        iv = get_random_bytes(16) # IV aléatoire indispensable pour CBC
        
        # 2. Lecture du fichier original
        with open(nom_fichier, 'rb') as f:
            data_originale = f.read()
            
        # 3. Chiffrement (AES-CBC + Padding)
        cipher = AES.new(key, AES.MODE_CBC, iv)
        data_chiffree = cipher.encrypt(pad(data_originale, AES.block_size))
        
        # 4. Écriture : On sauvegarde l'IV *suivi* des données chiffrées
        nom_sortie = nom_fichier + ".enc"
        with open(nom_sortie, 'wb') as f:
            f.write(iv)             # Les 16 premiers octets sont l'IV
            f.write(data_chiffree)  # Le reste est le message
            
        print(f" Succès ! Fichier chiffré créé : {nom_sortie}")
        return nom_sortie

    except FileNotFoundError:
        print(f" Erreur : Le fichier '{nom_fichier}' est introuvable.")
    except Exception as e:
        print(f" Erreur inattendue : {e}")

def dechiffrer_fichier(nom_fichier_chiffre, mot_de_passe):
    """Déchiffre un fichier .enc et retire l'extension"""
    try:
        # 1. Génération de la clé
        key = generer_cle_depuis_mdp(mot_de_passe)
        
        # 2. Lecture du fichier chiffré
        with open(nom_fichier_chiffre, 'rb') as f:
            iv = f.read(16)        # On récupère l'IV (les 16 premiers octets)
            data_chiffree = f.read() # Le reste
            
        # 3. Déchiffrement
        cipher = AES.new(key, AES.MODE_CBC, iv)
        try:
            data_dechiffree = unpad(cipher.decrypt(data_chiffree), AES.block_size)
        except ValueError:
            print(" Erreur : Mot de passe incorrect ou fichier corrompu (Padding invalide).")
            return None

        # 4. Écriture du fichier déchiffré
        # On enlève ".enc" et on ajoute "decrypted_" pour ne pas écraser l'original
        nom_base = os.path.splitext(nom_fichier_chiffre)[0]
        # Si le fichier avait un chemin, on garde juste le nom de fichier
        nom_base = os.path.basename(nom_base) 
        nom_sortie = "decrypted_" + nom_base
        
        with open(nom_sortie, 'wb') as f:
            f.write(data_dechiffree)
            
        print(f"✅ Succès ! Fichier déchiffré : {nom_sortie}")
        return nom_sortie

    except FileNotFoundError:
        print(f" Erreur : Le fichier '{nom_fichier_chiffre}' est introuvable.")
        return None

def verifier_fichiers(fichier1, fichier2):
    """Compare le contenu binaire de deux fichiers pour voir s'ils sont identiques"""
    print(f"🔍 Comparaison de '{fichier1}' et '{fichier2}'...")
    try:
        with open(fichier1, 'rb') as f1, open(fichier2, 'rb') as f2:
            contenu1 = f1.read()
            contenu2 = f2.read()
            
            if contenu1 == contenu2:
                print(" RÉSULTAT : Les fichiers sont STRICTEMENT IDENTIQUES.")
                print("   (Le chiffrement et le déchiffrement ont fonctionné parfaitement)")
            else:
                print(" RÉSULTAT : Les fichiers sont DIFFÉRENTS.")
                print("   (Il y a eu un problème lors du processus)")
                
    except FileNotFoundError as e:
        print(f" Erreur : Impossible de trouver un des fichiers ({e.filename}).")

# --- INTERFACE UTILISATEUR (MENU) ---

def main():
    while True: # Boucle pour pouvoir faire plusieurs actions de suite
        print("\n=== APPLICATION DE CHIFFREMENT AES-256 ===")
        print("1. Chiffrer un fichier")
        print("2. Déchiffrer un fichier")
        print("3. Vérifier que deux fichiers sont identiques")
        print("4. Quitter")
        
        choix = input("Votre choix (1-4) : ")

        if choix == '1':
            fichier = input("Nom du fichier à chiffrer (ex: oiseaux.bmp) : ")
            mdp = input("Choisissez un mot de passe : ")
            chiffrer_fichier(fichier, mdp)
            
        elif choix == '2':
            fichier = input("Nom du fichier chiffré (ex: oiseaux.bmp.enc) : ")
            mdp = input("Entrez le mot de passe original : ")
            dechiffrer_fichier(fichier, mdp)
        
        elif choix == '3':
            f1 = input("Nom du fichier original (ex: oiseaux.bmp) : ")
            f2 = input("Nom du fichier déchiffré (ex: decrypted_oiseaux.bmp) : ")
            verifier_fichiers(f1, f2)
            
        elif choix == '4':
            print("Au revoir !")
            break
            
        else:
            print("Choix invalide.")

# Pour lancer l'application directement dans le notebook/console :
if __name__ == "__main__":
    # On crée un petit fichier test pour la démo si besoin
    if not os.path.exists("test_secret.txt"):
        with open("test_secret.txt", "w") as f:
            f.write("Ceci est une donnée confidentielle pour le TP Crypto.")
    
    main()
```

```python
# ==========================================
# 1. CÔTÉ ALICE (L'Émetteur)
# ==========================================
def protocole_envoi_alice(message_clair, destinataire_pub_key):
    print("\n--- 1. Alice prépare l'envoi ---")
    
    # A. Créer une clé AES temporaire (Session Key)
    # On choisit une clé de 16 octets (128 bits) ou 32 (256 bits)
    session_key = get_random_bytes(32) 
    print("clé AES",session_key)
    
    # B. Chiffrer la clé AES avec la clé RSA publique de Bob
    # C'est l'étape "Enveloppe Numérique"
    cipher_rsa = PKCS1_OAEP.new(destinataire_pub_key)
    enc_session_key = cipher_rsa.encrypt(session_key)
    
    # C. Chiffrer le vrai message avec la clé AES
    cipher_aes = AES.new(session_key, AES.MODE_CBC)
    ciphertext = cipher_aes.encrypt(pad(message_clair, AES.block_size))
    
    # D. Alice emballe tout dans un seul paquet
    # Structure : [Taille Clé RSA] + [Clé AES Chiffrée] + [IV] + [Message]
    # Note : RSA 2048 bits produit toujours un bloc de 256 octets.
    packet = enc_session_key + cipher_aes.iv + ciphertext
    
    print(f"Alice envoie un paquet de {len(packet)} octets.",packet)
    return packet

# ==========================================
# 2. CÔTÉ BOB (Le Récepteur)
# ==========================================
def protocole_reception_bob(paquet_recu, ma_cle_privee):
    print("\n--- 2. Bob reçoit le paquet ---")
    
    # A. On découpe le paquet pour retrouver les morceaux
    # RSA 2048 bits = 256 octets pour la clé chiffrée
    rsa_key_length = ma_cle_privee.size_in_bytes() 
    
    enc_session_key = paquet_recu[:rsa_key_length]
    iv = paquet_recu[rsa_key_length : rsa_key_length + 16] # IV fait toujours 16 octets
    ciphertext = paquet_recu[rsa_key_length + 16:]
    
    try:
        # B. Bob déchiffre la clé AES avec sa clé RSA privée
        cipher_rsa = PKCS1_OAEP.new(ma_cle_privee)
        session_key = cipher_rsa.decrypt(enc_session_key)
        print(" Clé AES récupérée avec succès.")
        
        # C. Bob utilise la clé AES récupérée pour déchiffrer le message
        cipher_aes = AES.new(session_key, AES.MODE_CBC, iv)
        message_dechiffre = unpad(cipher_aes.decrypt(ciphertext), AES.block_size)
        
        return message_dechiffre
        
    except Exception as e:
        print(f" Erreur de déchiffrement : {e}")
        return None
```

```python
import random

# =========================================================
# PARTIE (C) : L'EXPONENTIELLE MODULAIRE
# Calculer (base^exp) % mod très vite
# =========================================================
def exp_modulaire(base, exp, mod):
    resultat = 1
    while exp > 0:
        # Si l'exposant est impair
        if exp % 2 == 1:
            resultat = (resultat * base) % mod
        # On met la base au carré
        base = (base * base) % mod
        # On divise l'exposant par 2
        exp = exp // 2
    return resultat
```

```python
# =========================================================
# PARTIE (A) : TEST DE PRIMALITÉ (Miller-Rabin Simplifié)
# Vérifie si un nombre 'n' est probablement premier
# =========================================================
def est_premier_miller_rabin(n, k=5):
    if n <= 1: return False
    if n <= 3: return True
    if n % 2 == 0: return False

    # On écrit n-1 sous la forme 2^r * d
    r, d = 0, n - 1
    while d % 2 == 0:
        r += 1
        d //= 2

    # On fait k tours de tests
    for _ in range(k):
        a = random.randint(2, n - 2)
        x = exp_modulaire(a, d, n) # Utilisation de notre fonction (c)
        
        if x == 1 or x == n - 1:
            continue
        
        for _ in range(r - 1):
            x = exp_modulaire(x, 2, n)
            if x == n - 1:
                break
        else:
            return False # N'est pas premier
            
    return True # Est probablement premier
```

```python
def generer_nombre_premier(nb_bits):
    """Génère un nombre premier aléatoire de N bits"""
    while True:
        # On génère un nombre impair aléatoire
        n = random.getrandbits(nb_bits)
        if n % 2 == 0: 
            n += 1
        # On teste s'il est premier
        if est_premier_miller_rabin(n):
            return n
```

```python
# =========================================================
# PARTIE (B) : EUCLIDE ÉTENDU (Pour trouver l'inverse)
# Trouve d tel que (d * e) % phi == 1
# =========================================================
def euclide_etendu(a, b):
    if a == 0:
        return b, 0, 1
    else:
        pgcd, x, y = euclide_etendu(b % a, a)
        return pgcd, y - (b // a) * x, x

def inverse_modulaire(e, phi):
    pgcd, x, y = euclide_etendu(e, phi)
    if pgcd != 1:
        raise Exception('L\'inverse modulaire n\'existe pas (e et phi ne sont pas premiers entre eux)')
    else:
        return x % phi
```

```python
# =========================================================
# LE COEUR DE RSA : GÉNÉRATION DES CLÉS
# =========================================================
def generer_cles_rsa_maison(nb_bits=16):
    print(f"--- Génération des clés ({nb_bits} bits) ---")
    
    # 1. Choisir deux nombres premiers p et q
    p = generer_nombre_premier(nb_bits)
    q = generer_nombre_premier(nb_bits)
    while p == q: # On s'assure qu'ils sont différents
        q = generer_nombre_premier(nb_bits)
    
    print(f"1. Premiers trouvés : p={p}, q={q}")

    # 2. Calculer le module n et l'indicateur d'Euler phi
    n = p * q
    phi = (p - 1) * (q - 1)
    print(f"2. Module n={n}, Phi={phi}")

    # 3. Choisir l'exposant public 'e'
    # e doit être premier avec phi. 65537 est un standard, mais on peut chercher plus petit.
    e = 65537
    # Si par hasard e n'est pas premier avec phi (très rare), on en cherche un autre
    while True:
        pgcd, _, _ = euclide_etendu(e, phi)
        if pgcd == 1: break
        e -= 2 # On essaie un autre impair
        
    print(f"3. Clé Publique (e) : {e}")

    # 4. Calculer l'exposant privé 'd' (L'inverse modulaire)
    d = inverse_modulaire(e, phi)
    print(f"4. Clé Privée (d)   : {d}")
    
    return ((e, n), (d, n))
```

```python
from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256

# --- 1. PRÉPARATION ---
key = RSA.generate(2048)
private_key = key
public_key = key.publickey()

message = b"Je certifie que ce TP est termine et valide."
print(f"Message original : {message}")

# --- 2. SIGNATURE (Côté Émetteur) ---
# A. On calcule le HASH du message
h = SHA256.new(message)
print(f"Empreinte (Hash) : {h.hexdigest()}")

# B. On signe le HASH avec la clé PRIVÉE
signature = pkcs1_15.new(private_key).sign(h)
print(f"Signature binaire générée (taille : {len(signature)} octets)")

# --- 3. VÉRIFICATION (Côté Récepteur) ---
# Le récepteur a reçu le message et la signature.
try:
    # A. Le récepteur recalcule le hash du message reçu
    h_verif = SHA256.new(message)
    
    # B. Il vérifie la signature avec la clé PUBLIQUE
    pkcs1_15.new(public_key).verify(h_verif, signature)
    
    print(" Signature VALIDE : Le message est authentique et intègre.")
    
except (ValueError, TypeError):
    print(" Signature INVALIDE : Le message a été modifié ou la clé est fausse.")

# --- 4. TEST DE FRAUDE (Optionnel) ---
print("\n--- Tentative de fraude ---")
faux_message = b"Je certifie que ce TP est termine et valide... PAS DU TOUT"
h_faux = SHA256.new(faux_message)
try:
    pkcs1_15.new(public_key).verify(h_faux, signature)
    print("Succès de la fraude (Impossible normalement)")
except:
    print(" Le système a détecté la modification du message !")
```

```python
import random
import hashlib
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad

# --- ETAPE 1 : Configuration Publique (Le "Jaune") ---
# Dans la réalité, ces nombres sont gigantesques et standardisés (RFC 3526)
# Pour le TP, on prend un petit nombre premier pour voir les calculs.
p = 23   # Module (Nombre premier)
g = 5    # Générateur (Base)

print(f"--- Configuration Publique ---")
print(f"Module p = {p}")
print(f"Générateur g = {g}")

# --- ETAPE 2 : Génération des Secrets (Les couleurs secrètes) ---
# Alice choisit un nombre secret 'a'
a_private = random.randint(1, p-1)

# Bob choisit un nombre secret 'b'
b_private = random.randint(1, p-1)

print(f"\nAlice choisit son secret : {a_private}")
print(f"Bob choisit son secret   : {b_private}")

# --- ETAPE 3 : Calcul des valeurs publiques (Le mélange à envoyer) ---
# Alice calcule A = g^a mod p
A_public = pow(g, a_private, p)

# Bob calcule B = g^b mod p
B_public = pow(g, b_private, p)

print(f"\n--- Echange sur le réseau (Public) ---")
print(f"Alice envoie A = {A_public}")
print(f"Bob envoie   B = {B_public}")

# --- ETAPE 4 : Calcul du Secret Partagé Final ---
# Alice reçoit B et calcule : S = B^a mod p
secret_alice = pow(B_public, a_private, p)

# Bob reçoit A et calcule : S = A^b mod p
secret_bob = pow(A_public, b_private, p)

print(f"\n--- Calcul Final ---")
print(f"Secret calculé par Alice : {secret_alice}")
print(f"Secret calculé par Bob   : {secret_bob}")

if secret_alice == secret_bob:
    print(" MAGIE ! Ils ont le même nombre sans se l'être dit.")
else:
    print(" Erreur de calcul.")
```

```python
# --- ETAPE 5 : Utilisation pour Chiffrement AES ---

# 1. Dérivation de la clé (Hashage du secret commun)
# On transforme l'entier secret en octets pour le hacher
secret_bytes = str(secret_alice).encode('utf-8')
cle_aes_commune = hashlib.sha256(secret_bytes).digest()

print(f"\nClé AES dérivée du secret : {cle_aes_commune.hex()}")

# 2. Simulation d'une communication sécurisée
message_alice = b"Coucou Bob, personne ne peut nous lire !"

# Alice chiffre avec la clé dérivée
cipher_alice = AES.new(cle_aes_commune, AES.MODE_CBC)
msg_chiffre = cipher_alice.encrypt(pad(message_alice, AES.block_size))
iv = cipher_alice.iv

print(f"Message chiffré envoyé : {msg_chiffre.hex()}")

# 3. Bob déchiffre avec SA clé dérivée (qui est la même)
cipher_bob = AES.new(cle_aes_commune, AES.MODE_CBC, iv)
msg_recu = unpad(cipher_bob.decrypt(msg_chiffre), AES.block_size)

print(f"Message lu par Bob : {msg_recu.decode('utf-8')}")
```
